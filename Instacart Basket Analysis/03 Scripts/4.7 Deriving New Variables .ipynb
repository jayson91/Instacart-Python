## If-Statements with User-Defined Functions

```python
#Define a function

def price_label (row):
    if row['prices'] <=5:
        return "Low-range product"
    elif (row["prices"]>5 and row["prices"]<=15):
         return "Mid-range product"
    elif row["prices"] >15:
        return "High-range"
    else: return "Not enough data"
```

```python
#Apply the function

df['price_range']= df.apply(price_label,axis=1)
```

```python
# Check the frequency of new column 
df['price_range'].value_counts(dropna= False)
```

```python
#Check the max value from the subset data
df['prices'].max()
```

## If-Statements with the loc() Function

```python
#using loc function create new column andlabel them based on prices
df.loc[df['prices']>15, "price_range_loc"]= "High-range product"
df.loc[(df['prices']>5) &(df['prices']<=15), "price_range_loc"] = "Mid-range product"
df.loc[df["prices"]<=5, "price_range_loc"]= "Low-range product"
```

```python
#Check the frequency of the new column
df["price_range_loc"].value_counts(dropna= False)
```

### Task
#### 2.) Suppose your clients have changed their minds about the labels you created in your “busiest_day” column. Now, they want “Busiest day” to become “Busiest days” (plural). This label should correspond with the two busiest days of the week as opposed to the single busiest day. At the same time, they’d also like to know the two slowest days. Create a new column for this using a suitable method. 

```python
#For loop used to label the order day of the week
result_2 = []

for x in ords_prods_merged["orders_day_of_week"]:
    if x==0 or x==1:
        result_2.append("Busiest days")
    elif x==4 or x==3:
        result_2.append("Least days")
    else:
        result_2.append("Regularly busy")
```

```python
result_2
```

```python
ords_prods_merged["busiest_days"]=result_2
```

```python
ords_prods_merged.columns
```

```python
ords_prods_merged.head()
```

```python
#frequency for this new column.
ords_prods_merged['busiest_days'].value_counts(dropna=False)
```

##### When checked with our original frequency counts, these values add up to the sums of each "grouping". ie - the total for busiest days equals the sum of value 0 and value 1, and the values for least busy days adds to the sum of value 3 and 4, confirming that the loop operated correctly

#### 4.) When too many users make Instacart orders at the same time, the app freezes. The senior technical officer at Instacart wants you to identify the busiest hours of the day. Rather than by hour, they want periods of time labeled “Most orders,” “Average orders,” and “Fewest orders.” Create a new column containing these labels called “busiest_period_of_day.”

```python
#frequency for each hour of the day
ords_prods_merged["order_hour_of_day"].value_counts()
```

```python
#For loop used for the busiest hour of the day 
result3=[]

for x in ords_prods_merged["order_hour_of_day"]:
    if x in [10,11,14,15,13,12,16,9]:
        result3.append("Most orders")
    elif x in [3,4,2,5,1,0,6,23]:
        result3.append("Fewest orders")
    else:
        result3.append("Average orders")
```

```python
#Check the result3
result3
```

```python
#frequency for this new column.
ords_prods_merged["busiest_hours"]=result3
ords_prods_merged["busiest_hours"].value_counts()
```
