### 3. The Instacart officers are interested in comparing customer behavior in different geographic areas. Create a regional segmentation of the data. You’ll need to create a “Region” column based on the “State” column from your customers data set. 

```python
# create a for loop where state's region is mentioned and stored in result

result = []

for x in ords_prods_cust_merge["state"]:
    if x in ["Connecticut","Maine", "Massachusetts", "New Hampshire", "New Jersey",
               "Rhode Island", "Vermont", "New York", "Pennsylvania" ]:
        result.append ("Northeast")
        
    elif x in ["Illinois", "Indiana", "Iowa", "Michigan", "Minnesota", "Missouri", 
               "Nebraska", "North Dakota" ,"Ohio", "South Dakota", "Wisconsin"]:
        result.append ("Midwest")
            
    elif x in ["Alabama", "Arkansas", "Delaware", "District of Columbia", "Florida", "Georgia", "Kansas", "Kentucky",
               "Louisiana", "Maryland", "Mississippi", "North Carolina", "Oklahoma", "South Carolina", "Tennessee",
               "Texas", "Virginia", "West Virginia"]:
        result.append("South")
    
    elif x in ["Alaska", "Arizona", "California", "Colorado", "Hawaii", "Idaho", "Montana", "Nevada", "New Mexico",
               "Oregon", "Utah", "Washington", "Wyoming"]:
        result.append("West")
    
    else:
        result.append("Unknown")
    
```

```python
result
```

```python
# Save the result in new column called region

ords_prods_cust_merge["region"]=result
```

```python
# Check the value counts Region

ords_prods_cust_merge["region"].value_counts(dropna= False)
```
